```python
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import BertForSequenceClassification, AutoTokenizer
from torch.optim import AdamW
from sklearn.model_selection import train_test_split
import numpy as np
from tqdm import tqdm
import pandas as pd
from transformers import get_linear_schedule_with_warmup
from sklearn.metrics import cohen_kappa_score, confusion_matrix

# 1. Préparation des données améliorées
# --------------------------------------

class NotationDataset(Dataset):
    def __init__(self, questions, reponses, notes, tokenizer, max_length=512):
        self.questions = questions
        self.reponses = reponses
        self.notes = notes
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.questions)

    def __getitem__(self, idx):
        texte = self.questions[idx] + self.tokenizer.sep_token + self.reponses[idx]

        encodings = self.tokenizer(
            texte,
            truncation=True,
            padding='max_length',
            max_length=self.max_length,
            return_tensors='pt'
        )

        return {
            'input_ids': encodings['input_ids'].squeeze(),
            'attention_mask': encodings['attention_mask'].squeeze(),
            'labels': torch.tensor(self.notes[idx], dtype=torch.long)
        }

def creer_donnees_islamique_arabe(nb_exemples=1000):
    # Thèmes de l'éducation islamique
    themes_islamiques = [
        "أركان الإسلام",                   # Les piliers de l'Islam
        "أركان الإيمان",                   # Les piliers de la foi
        "سيرة النبي محمد",                 # La biographie du prophète Muhammad
        "مفهوم الزكاة وأهميتها",          # Le concept de la Zakat et son importance
        "الحج ومناسكه",                   # Le pèlerinage et ses rituels
        "آداب التعامل في الإسلام",         # L'éthique des relations en Islam
        "مفهوم الصوم وفوائده",            # Le concept du jeûne et ses bienfaits
        "دور المسجد في المجتمع الإسلامي"   # Le rôle de la mosquée dans la société islamique
    ]

    # Thèmes de la langue arabe
    themes_arabes = [
        "قواعد النحو العربي",              # Les règles de grammaire arabe
        "البلاغة العربية وأقسامها",        # La rhétorique arabe et ses divisions
        "الشعر العربي في العصر الجاهلي",    # La poésie arabe à l'époque préislamique
        "المعلقات السبع وأهميتها",         # Les sept poèmes suspendus et leur importance
        "أدب المقامات في الأدب العربي",     # La littérature des Maqamat dans la littérature arabe
        "تطور الرواية العربية",            # L'évolution du roman arabe
        "تأثير اللغة العربية في اللغات الأخرى", # L'influence de la langue arabe sur d'autres langues
        "إعراب الجملة العربية"             # L'analyse grammaticale de la phrase arabe
    ]

    # Combiner tous les thèmes
    tous_themes = themes_islamiques + themes_arabes

    questions = []
    reponses = []
    notes = []

    for i in range(nb_exemples):
        theme_idx = i % len(tous_themes)
        theme = tous_themes[theme_idx]

        # Types de questions pour diversifier
        if i % 4 == 0:
            question = f"اشرح مفهوم {theme} وأهميته."  # Expliquer le concept et son importance
        elif i % 4 == 1:
            question = f"ما هي العناصر الأساسية ل{theme}؟"  # Quels sont les éléments fondamentaux de...
        elif i % 4 == 2:
            question = f"قارن بين {theme} و{tous_themes[(theme_idx + 1) % len(tous_themes)]}."  # Comparer...
        else:
            question = f"اذكر تطبيقات {theme} في حياتنا اليومية."  # Mentionner les applications dans notre vie quotidienne

        questions.append(question)

        # Générer des réponses de qualité différenciée pour chaque note
        qualite = i % 5  # 0-4 (correspond aux notes 1-5)

        if theme in themes_islamiques:
            # Réponses pour les thèmes islamiques
            if qualite == 0:  # Très mauvaise (1/5)
                reponse = f"{theme} مهم في الإسلام."  # Réponse minimale, sans contenu réel

            elif qualite == 1:  # Mauvaise (2/5)
                reponse = f"{theme} من الأمور المهمة في الدين الإسلامي. يجب على كل مسلم معرفة ذلك."
                # Réponse générique avec peu d'informations concrètes

            elif qualite == 2:  # Moyenne (3/5)
                reponse = f"{theme} يعتبر من المفاهيم الأساسية في الإسلام. "
                reponse += f"وقد ورد ذكره في القرآن والسنة. من فوائده أنه يساعد المسلم على تقوية إيمانه. "
                reponse += f"كما أنه يساهم في بناء المجتمع الإسلامي."
                # Réponse avec informations basiques mais sans profondeur

            elif qualite == 3:  # Bonne (4/5)
                reponse = f"{theme} مفهوم أساسي في الدين الإسلامي ويرتبط ارتباطاً وثيقاً بحياة المسلم. "
                reponse += f"ورد ذكره في القرآن الكريم في عدة آيات منها قوله تعالى: '...' وقال النبي صلى الله عليه وسلم: '...'. "
                reponse += f"يتضمن {theme} عدة جوانب مهمة: أولاً، الجانب الروحي الذي يتمثل في... ثانياً، الجانب الاجتماعي الذي يظهر في... "
                reponse += f"ويساهم تطبيق {theme} في تعزيز التكافل الاجتماعي وتقوية الروابط بين أفراد المجتمع."
                # Réponse structurée avec références et analyse

            else:  # Excellente (5/5)
                reponse = f"{theme} ركن أساسي من أركان الدين الإسلامي ومفهوم محوري في بناء شخصية المسلم والمجتمع الإسلامي. "
                reponse += f"ورد ذكره في القرآن الكريم في مواضع عديدة، منها قوله تعالى: '...' في سورة ... الآية ... وقوله تعالى: '...' في سورة ... الآية ... "
                reponse += f"وقد أكد النبي محمد صلى الله عليه وسلم على أهميته في أحاديث كثيرة منها: '...' (رواه البخاري)، وقوله: '...' (رواه مسلم). "
                reponse += f"يمكن تقسيم {theme} إلى عدة أقسام رئيسية: "
                reponse += f"القسم الأول: ... ويشمل ... ويهدف إلى ... والدليل عليه قوله تعالى: '...'. "
                reponse += f"القسم الثاني: ... وفيه ... ومن حكمه ... كما قال العلماء: '...'. "
                reponse += f"القسم الثالث: ... ويتضمن ... وفائدته ... وقد أجمع الفقهاء على أن ... "
                reponse += f"من آثار {theme} على الفرد: تزكية النفس، وتقوية الإيمان، وتهذيب السلوك. "
                reponse += f"ومن آثاره على المجتمع: تعزيز التكافل الاجتماعي، وترسيخ القيم الأخلاقية، ونشر المحبة والتعاون. "
                reponse += f"ويمكن تطبيق {theme} في حياتنا المعاصرة من خلال ... و... و... "
                reponse += f"وفي الختام، يعد {theme} من أهم الركائز التي يقوم عليها المجتمع الإسلامي المتوازن، ولا غنى للمسلم عن فهمه وتطبيقه في حياته اليومية."
                # Réponse complète, détaillée et structurée avec références, analyse et application

        else:  # Thèmes de la langue arabe
            if qualite == 0:  # Très mauvaise (1/5)
                reponse = f"{theme} مهم في اللغة العربية."  # Réponse minimale

            elif qualite == 1:  # Mauvaise (2/5)
                reponse = f"{theme} من المواضيع المهمة في اللغة العربية. يجب على كل طالب دراسته."
                # Réponse générique sans contenu substantiel

            elif qualite == 2:  # Moyenne (3/5)
                reponse = f"{theme} هو أحد المباحث المهمة في علوم اللغة العربية. "
                reponse += f"ظهر في العصر ... وله عدة قواعد أساسية. من أهميته أنه يساعد على فهم النصوص العربية بشكل صحيح."
                # Informations de base mais sans exemples ou analyses détaillées

            elif qualite == 3:  # Bonne (4/5)
                reponse = f"{theme} من المباحث الأساسية في علوم اللغة العربية، وقد نشأ في العصر ... على يد العالم ... "
                reponse += f"من أهم قواعد {theme}: القاعدة الأولى التي تنص على ... والقاعدة الثانية المتعلقة بـ ... "
                reponse += f"ومن أمثلة ذلك: ... و... و... "
                reponse += f"أهمية {theme} تكمن في أنه يساعد على فهم النصوص العربية وإنتاجها بشكل سليم، كما يسهم في الحفاظ على جمالية اللغة العربية وبلاغتها."
                # Réponse structurée avec exemples et analyse de l'importance

            else:  # Excellente (5/5)
                reponse = f"{theme} من أهم المباحث في علوم اللغة العربية وأكثرها دقة وتفصيلاً. "
                reponse += f"نشأ {theme} في العصر ... على يد العالم ... ثم تطور على مر العصور بإسهامات علماء مثل ... و... و... "
                reponse += f"يُقسم {theme} إلى عدة أقسام رئيسية: "
                reponse += f"القسم الأول: ... ويختص بـ ... ومن أمثلته قول الشاعر: '...' "
                reponse += f"والقسم الثاني: ... ويتناول ... ومثاله في النثر العربي: '...' "
                reponse += f"أما القسم الثالث فهو: ... ويتميز بـ ... ونجد مثالاً عليه في القرآن الكريم: '...' "
                reponse += f"من القواعد الأساسية في {theme}: القاعدة الأولى التي تنص على ... والاستثناء منها ... "
                reponse += f"والقاعدة الثانية المتعلقة بـ ... وتطبيقاتها في ... "
                reponse += f"القاعدة الثالثة الخاصة بـ ... وأهميتها في ... "
                reponse += f"اختلف علماء اللغة في بعض مسائل {theme} مثل مسألة ... حيث يرى البصريون ... بينما يرى الكوفيون ... "
                reponse += f"أهمية {theme} تتجلى في عدة جوانب: لغوياً، يساعد على ... وأدبياً، يسهم في ... وثقافياً، يعزز ... "
                reponse += f"من تطبيقات {theme} في العصر الحديث: استخدامه في ... وتوظيفه في ... ودوره في ... "
                reponse += f"وفي الختام، يبقى {theme} ركيزة أساسية من ركائز اللغة العربية التي تحفظ أصالتها وتضمن استمراريتها وتطورها، ويجب على دارسي اللغة العربية إتقانه وفهم قواعده وتطبيقاتها المختلفة."
                # Réponse complète avec contexte historique, classifications, exemples littéraires, débats, et applications

        reponses.append(reponse)
        notes.append(qualite)

    return questions, reponses, notes

# Fonction pour augmenter les données
def augmenter_donnees(questions, reponses, notes):
    """Augmente les données en ajoutant des variations aux réponses."""
    questions_aug = []
    reponses_aug = []
    notes_aug = []

    for i, (question, reponse, note) in enumerate(zip(questions, reponses, notes)):
        # Ne pas augmenter toutes les données, seulement certaines
        if i % 3 != 0:  # Augmenter environ 2/3 des données
            continue

        # Si la réponse est suffisamment longue
        if len(reponse.split()) > 20:
            # 1. Inverser l'ordre des paragraphes
            paragraphes = reponse.split('.')
            if len(paragraphes) > 3:
                milieu = len(paragraphes) // 2
                reponse_aug = '. '.join(paragraphes[milieu:] + paragraphes[:milieu])
                questions_aug.append(question)
                reponses_aug.append(reponse_aug)
                # Légèrement réduire la note car l'ordre est modifié
                notes_aug.append(max(0, note-1) if note > 2 else note)

            # 2. Ajouter des fautes typographiques mineures pour les bonnes réponses
            if note >= 3:  # Uniquement pour les bonnes et excellentes réponses
                # Simuler quelques erreurs typographiques
                mots = reponse.split()
                if len(mots) > 10:
                    # Supprimer aléatoirement 5-10% des mots
                    indices_a_supprimer = np.random.choice(
                        range(len(mots)),
                        size=max(1, int(len(mots) * 0.05)),
                        replace=False
                    )
                    mots_modifies = [mot for i, mot in enumerate(mots) if i not in indices_a_supprimer]
                    reponse_aug = ' '.join(mots_modifies)
                    questions_aug.append(question)
                    reponses_aug.append(reponse_aug)
                    # Réduire la note car il manque des mots
                    notes_aug.append(max(0, note-1))

    # Combiner les données originales et augmentées
    questions_complet = questions + questions_aug
    reponses_complet = reponses + reponses_aug
    notes_complet = notes + notes_aug

    print(f"Augmentation de données: {len(questions)} → {len(questions_complet)} exemples")

    return questions_complet, reponses_complet, notes_complet

# Fonction pour geler certaines couches de BERT
def freeze_bert_layers(model, num_layers_to_freeze=8):
    """Gèle les premières couches de BERT pour un fine-tuning plus efficace."""
    # Gel des embeddings
    for param in model.bert.embeddings.parameters():
        param.requires_grad = False

    # Gel des premières couches d'encodeur
    for layer in model.bert.encoder.layer[:num_layers_to_freeze]:
        for param in layer.parameters():
            param.requires_grad = False

    return model

# Fonction d'évaluation détaillée
def evaluate_detailed(model, dataloader, device):
    model.eval()
    all_preds = []
    all_labels = []
    total_loss = 0

    with torch.no_grad():
        for batch in tqdm(dataloader, desc="Evaluating"):
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            total_loss += loss.item()

            predictions = torch.argmax(outputs.logits, dim=1)

            all_preds.extend(predictions.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Calcul de l'exactitude
    accuracy = sum(1 for p, l in zip(all_preds, all_labels) if p == l) / len(all_preds)

    # Calcul du Kappa de Cohen (mesure d'accord inter-annotateurs)
    kappa = cohen_kappa_score(all_labels, all_preds)

    # Matrice de confusion
    conf_matrix = confusion_matrix(all_labels, all_preds)

    # Écart moyen absolu (à quel point les prédictions sont loin de la réalité)
    mae = sum(abs(p - l) for p, l in zip(all_preds, all_labels)) / len(all_preds)

    return {
        "loss": total_loss / len(dataloader),
        "accuracy": accuracy,
        "kappa": kappa,
        "confusion_matrix": conf_matrix,
        "mae": mae
    }

# Fonction principale d'entraînement
def train_improved_model():
    # 1. Charger le tokenizer et le modèle
    print("Chargement du tokenizer et du modèle...")
    tokenizer = AutoTokenizer.from_pretrained("aubmindlab/bert-base-arabertv2")
    model = BertForSequenceClassification.from_pretrained(
        "aubmindlab/bert-base-arabertv2",
        num_labels=5  # Pour la notation de 1 à 5
    )

    # 2. Créer les données d'entraînement
    print("Création des données d'entraînement avec thèmes islamiques et arabes...")
    questions, reponses, notes = creer_donnees_islamique_arabe(2000)

    # 3. Augmenter les données
    print("Augmentation des données...")
    questions, reponses, notes = augmenter_donnees(questions, reponses, notes)

    # 4. Division stratifiée des données
    print("Division des données en ensembles d'entraînement et de validation...")
    q_train, q_val, r_train, r_val, n_train, n_val = train_test_split(
        questions, reponses, notes, test_size=0.2, random_state=42, stratify=notes
    )

    print(f"Répartition des notes dans l'ensemble d'entraînement:")
    for note in range(5):
        count = n_train.count(note)
        print(f"Note {note+1}/5: {count} exemples ({count/len(n_train)*100:.1f}%)")

    # 5. Créer les datasets
    train_dataset = NotationDataset(q_train, r_train, n_train, tokenizer)
    val_dataset = NotationDataset(q_val, r_val, n_val, tokenizer)

    # 6. Créer les dataloaders avec une taille de batch optimisée
    batch_size = 16  # Augmenter si possible selon la mémoire GPU
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size)

    # 7. Geler certaines couches pour un fine-tuning plus efficace
    print("Gel des premières couches de BERT...")
    model = freeze_bert_layers(model, num_layers_to_freeze=6)

    # 8. Configurer l'optimiseur avec un learning rate adapté
    optimizer = AdamW(
        [p for p in model.parameters() if p.requires_grad],
        lr=3e-5,
        weight_decay=0.01
    )

    # 9. Configurer le scheduler de learning rate
    total_steps = len(train_loader) * 15  # Pour 15 époques
    scheduler = get_linear_schedule_with_warmup(
        optimizer,
        num_warmup_steps=int(total_steps * 0.1),  # 10% du total pour le warmup
        num_training_steps=total_steps
    )

    # 10. Déplacer le modèle sur GPU si disponible
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Utilisation de l'appareil: {device}")
    model.to(device)

    # 11. Entraînement avec early stopping
    num_epochs = 15
    best_val_acc = 0
    patience = 3
    patience_counter = 0
    history = {"train_loss": [], "train_acc": [], "val_loss": [], "val_acc": [], "val_kappa": []}

    print(f"Début de l'entraînement pour {num_epochs} époques maximum...")
    for epoch in range(num_epochs):
        print(f"Époque {epoch+1}/{num_epochs}")

        # Entraînement
        model.train()
        total_loss = 0
        correct = 0
        total = 0

        progress_bar = tqdm(train_loader, desc="Training")
        for batch in progress_bar:
            # Déplacer les tenseurs vers le GPU
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            # Réinitialiser les gradients
            optimizer.zero_grad()

            # Forward pass
            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss

            # Backward pass
            loss.backward()

            # Clip gradient norm pour éviter l'explosion des gradients
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

            # Mise à jour des poids
            optimizer.step()
            scheduler.step()

            # Statistiques
            total_loss += loss.item()

            # Calcul de précision
            predictions = torch.argmax(outputs.logits, dim=1)
            correct += (predictions == labels).sum().item()
            total += labels.size(0)

            # Mettre à jour la barre de progression
            progress_bar.set_postfix({
                'loss': total_loss / (progress_bar.n + 1),
                'acc': correct / total,
                'lr': scheduler.get_last_lr()[0]
            })

        train_loss = total_loss / len(train_loader)
        train_acc = correct / total
        history["train_loss"].append(train_loss)
        history["train_acc"].append(train_acc)

        print(f"Training Loss: {train_loss:.4f}, Training Accuracy: {train_acc:.4f}")

        # Évaluation détaillée
        eval_results = evaluate_detailed(model, val_loader, device)
        val_loss = eval_results["loss"]
        val_acc = eval_results["accuracy"]
        val_kappa = eval_results["kappa"]
        val_mae = eval_results["mae"]

        history["val_loss"].append(val_loss)
        history["val_acc"].append(val_acc)
        history["val_kappa"].append(val_kappa)

        print(f"Validation Loss: {val_loss:.4f}, Validation Accuracy: {val_acc:.4f}")
        print(f"Kappa: {val_kappa:.4f}, MAE: {val_mae:.4f}")
        print("Matrice de confusion:")
        print(eval_results["confusion_matrix"])

        # Sauvegarder le meilleur modèle
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), "best_arabert_notation_islamique_arabe.pt")
            print(f"Meilleur modèle sauvegardé! Accuracy: {val_acc:.4f}")
            patience_counter = 0
        else:
            patience_counter += 1

        # Early stopping
        if patience_counter >= patience:
            print(f"Early stopping après {epoch+1} époques")
            break

        print("-" * 50)

    # 12. Charger le meilleur modèle pour les tests
    model.load_state_dict(torch.load("best_arabert_notation_islamique_arabe.pt"))

    return model, tokenizer, best_val_acc, history

# Fonction pour tester le modèle sur des exemples réels
def tester_model(model, tokenizer, device):
    # Exemples de test pour l'éducation islamique
    exemples_test_islamique = [
        {
            "question": "اشرح مفهوم الزكاة وأهميتها في الإسلام.",
            "reponse": "الزكاة هي الركن الثالث من أركان الإسلام، وهي فريضة مالية على كل مسلم تتوفر فيه شروط معينة. تُخرج الزكاة بنسبة 2.5% من المال إذا بلغ النصاب وحال عليه الحول. من أهم أهدافها تطهير المال والنفس، ومساعدة الفقراء والمحتاجين، وتحقيق التكافل الاجتماعي."
        },
        {
            "question": "ما هي أركان الإيمان؟",
            "reponse": "الإيمان بالله."
        },
        {
            "question": "اشرح مفهوم الصوم وفوائده.",
            "reponse": "الصوم هو الامتناع عن الطعام والشراب من الفجر إلى المغرب في شهر رمضان المبارك. من فوائده الصحية: تنظيف الجسم من السموم، وتقوية المناعة. ومن فوائده الروحية: تعزيز التقوى والصبر والإحساس بمعاناة الفقراء."
        }
    ]

    # Exemples de test pour la langue arabe
    exemples_test_arabe = [
        {
            "question": "اشرح مفهوم البلاغة العربية وأقسامها.",
            "reponse": "البلاغة العربية هي فن القول الجميل والتعبير المؤثر. تنقسم إلى ثلاثة أقسام رئيسية: علم المعاني الذي يهتم بمطابقة الكلام لمقتضى الحال، وعلم البيان الذي يهتم بطرق التعبير المختلفة عن المعنى الواحد، وعلم البديع الذي يهتم بتحسين الكلام. من أهم علماء البلاغة: عبد القاهر الجرجاني صاحب كتاب دلائل الإعجاز، والسكاكي صاحب مفتاح العلوم."
        },
        {
            "question": "ما هي المعلقات السبع وأهميتها؟",
            "reponse": "المعلقات شعر جاهلي."
        },
        {
            "question": "اشرح قواعد النحو العربي.",
            "reponse": "النحو العربي هو علم يبحث في أحوال أواخر الكلم إعراباً وبناءً. من أهم قواعده: الإعراب والبناء، والمرفوعات مثل المبتدأ والخبر والفاعل، والمنصوبات مثل المفعول به والحال، والمجرورات مثل المضاف إليه."
        }
    ]

    # Combiner tous les exemples de test
    tous_exemples = exemples_test_islamique + exemples_test_arabe

    resultats = []

    print("Test du modèle sur des exemples réels:")
    for i, exemple in enumerate(tous_exemples):
        question = exemple["question"]
        reponse = exemple["reponse"]

        # Préparation de l'entrée
        texte = question + tokenizer.sep_token + reponse
        inputs = tokenizer(
            texte,
            truncation=True,
            padding='max_length',
            max_length=512,
            return_tensors="pt"
        )

        # Déplacer les entrées vers l'appareil approprié
        input_ids = inputs['input_ids'].to(device)
        attention_mask = inputs['attention_mask'].to(device)

        # Prédiction
        model.eval()
        with torch.no_grad():
            outputs = model(input_ids=input_ids, attention_mask=attention_mask)

        # Obtenir la note prédite (0-4) et convertir à l'échelle 1-5
        note_predite = torch.argmax(outputs.logits, dim=1).item() + 1

        # Obtenir les probabilités pour chaque note
        probs = torch.softmax(outputs.logits, dim=1)[0].cpu().numpy()

        resultat = {
            "question": question,
            "reponse": reponse,
            "note": note_predite,
            "confiance": probs[note_predite-1],
            "distribution": {f"note_{i+1}": float(probs[i]) for i in range(5)}
        }

        resultats.append(resultat)

        # Afficher le résultat
        print(f"Exemple {i+1}:")
        print(f"Question: {question}")
        print(f"Réponse (extrait): {reponse[:100]}...")
        print(f"Note attribuée: {note_predite}/5 (confiance: {resultat['confiance']:.2f})")
        print("Distribution des probabilités:")
        for note, prob in resultat["distribution"].items():
            print(f"  {note}: {prob:.4f}")
        print("-" * 50)

    return resultats

# Fonction pour visualiser les résultats de l'entraînement
def visualiser_resultats(history):
    import matplotlib.pyplot as plt

    # Créer la figure avec 2 sous-graphiques
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Graphique des pertes
    ax1.plot(history["train_loss"], label="Train")
    ax1.plot(history["val_loss"], label="Validation")
    ax1.set_title("Évolution de la perte")
    ax1.set_xlabel("Époque")
    ax1.set_ylabel("Perte")
    ax1.legend()

    # Graphique des précisions
    ax2.plot(history["train_acc"], label="Train")
    ax2.plot(history["val_acc"], label="Validation")
    ax2.set_title("Évolution de la précision")
    ax2.set_xlabel("Époque")
    ax2.set_ylabel("Précision")
    ax2.legend()

    plt.tight_layout()
    plt.savefig("resultats_entrainement.png")
    plt.show()

# Fonction pour analyser les erreurs
def analyser_erreurs(model, tokenizer, dataloader, device):
    model.eval()
    erreurs = []

    with torch.no_grad():
        for batch in tqdm(dataloader, desc="Analyser les erreurs"):
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            predictions = torch.argmax(outputs.logits, dim=1)

            # Trouver les erreurs
            for i in range(len(labels)):
                if predictions[i] != labels[i]:
                    # Reconstruire le texte original
                    texte = tokenizer.decode(input_ids[i], skip_special_tokens=True)

                    erreurs.append({
                        "texte": texte,
                        "note_reelle": labels[i].item() + 1,  # +1 pour convertir de 0-4 à 1-5
                        "note_predite": predictions[i].item() + 1,
                        "confiance": torch.softmax(outputs.logits[i], dim=0)[predictions[i]].item()
                    })

    # Analyser les types d'erreurs les plus courants
    print(f"Nombre total d'erreurs: {len(erreurs)}")

    # Erreurs par écart de note
    ecarts = {}
    for erreur in erreurs:
        ecart = abs(erreur["note_reelle"] - erreur["note_predite"])
        ecarts[ecart] = ecarts.get(ecart, 0) + 1

    print("Distribution des erreurs par écart de note:")
    for ecart, count in sorted(ecarts.items()):
        print(f"Écart de {ecart} point(s): {count} erreurs ({count/len(erreurs)*100:.1f}%)")

    # Afficher quelques exemples d'erreurs
    print("\nExemples d'erreurs:")
    for i, erreur in enumerate(erreurs[:5]):
        print(f"Exemple {i+1}:")
        print(f"Texte: {erreur['texte'][:100]}...")
        print(f"Note réelle: {erreur['note_reelle']}, Note prédite: {erreur['note_predite']}")
        print(f"Confiance: {erreur['confiance']:.4f}")
        print("-" * 40)

    return erreurs

# Fonction principale pour exécuter tout le processus
def main():
    # Entraînement du modèle
    model, tokenizer, best_accuracy, history = train_improved_model()

    # Visualisation des résultats
    visualiser_resultats(history)

    # Test du modèle sur des exemples réels
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    resultats_test = tester_model(model, tokenizer, device)

    print(f"Meilleure précision obtenue: {best_accuracy:.4f}")

    # Sauvegarde du modèle final
    model.save_pretrained("arabert_notation_final")
    tokenizer.save_pretrained("arabert_notation_final")

    print("Modèle et tokenizer sauvegardés avec succès!")

    return model, tokenizer, history, resultats_test

# Exécution du code principal si exécuté comme script
if __name__ == "__main__":
    main()
```
